```python
from pyhht.emd import EMD
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pyhht.visualization import plot_imfs
```

```python
# 
namelist=['CMCSA-5year.csv','TSLA-5year.csv','MA-5year.csv','TOT-5year.csv','DUK-5year.csv','EXC-5year.csv'
         ,'GE-5year.csv','HD-5year.csv']
for name in namelist:
    stock_dir='../../dataset/5year/'+name
    dataset = pd.read_csv(open(stock_dir),header=0)
    dt_open=dataset['Close']
    data = dt_open.values
    s = data.ravel()
    decomposer = EMD(s)               
    #emd
    IMF = decomposer.decompose()
    #print(IMF.shape)
    imf_data = pd.DataFrame(IMF.T)
    imf_data.to_csv('../../dataset/emd/'+name[:-4]+'-emd.csv')
    plot_imfs(s,IMF)
```
